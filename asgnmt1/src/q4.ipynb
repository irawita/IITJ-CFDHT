```python
import numpy as np
import math

def sum_series(n):
    total = 0.0
    for k in range(1, n+1):
        total += 1/(k**2)
    return total


def find_n_terms(tol):
    analytical = (math.pi**2)/6
    
    n = 1
    total = 0.0
    
    while True:
        total += 1/(n**2)
        
        if abs(analytical - total) < tol:
            break
        
        n += 1
    
    return n, total
```

Here the analytical value of $\pi^2/6$ is taken as the limit for the infinite sum. The video contains a proof, which is also otherwise easily available online.  
https://www.youtube.com/watch?v=NgqST16ITyA
The tolerance for n decimal places is taken to be $0.5 * 10^{-n}$. $n1$ and $n2$ store the terms needed for $10^{-2}$ and $10^{-5}$ respectively.

```python
n1, y1 = find_n_terms(5e-3)

n2, y2 = find_n_terms(5e-6)

analytical = (math.pi**2)/6

print("y1:", y1, "n1:", n1)
print("y2:", y2, "n2:", n2)
print("Exact:", analytical)
```

```python
import matplotlib.pyplot as plt

analytical = (math.pi**2)/6

N = 500
partial_sums = []
errors = []

total = 0.0

for n in range(1, N+1):
    total += 1/(n**2)
    partial_sums.append(total)
    errors.append(abs(analytical - total))

x = np.arange(1, N+1)

plt.figure()
plt.plot(x, partial_sums, label="Partial Sum")
plt.axhline(analytical, linestyle='--', label="π²/6")
plt.xlabel("n")
plt.ylabel("Sum")
plt.title("Convergence of ∑ 1/n²")
plt.legend()
plt.grid()
plt.show()
```
