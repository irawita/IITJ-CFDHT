```python
def digit_sum(n):
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total


def generate_xi(m, p):
    xi = []
    
    start = 10**(m-1)
    end = 10**m
    
    for i in range(start, end):
        if digit_sum(i) % p == 0:
            xi.append(i)
    
    return xi


def compute_sums(xi):
    S = 0
    sumsq = 0
    
    for val in xi:
        S += val
        sumsq += val * val
    
    y1 = S * S
    y2 = y1 - sumsq
    y3 = y2 // 2
    
    return y1, y2, y3


def solve_case(m, p):
    xi = generate_xi(m, p)
    
    print(f"\nCase m={m}, p={p}")
    print("Count of xi:", len(xi))
    
    y1, y2, y3 = compute_sums(xi)
    
    print("y1:", y1)
    print("y2:", y2)
    print("y3:", y3)
    
    print("Check y2 == 2*y3:", y2 == 2*y3)
    print("Check y1 == y2 + sum(x_i^2):", y1 == y2 + sum(x*x for x in xi))


solve_case(3, 13)
solve_case(5, 9)
solve_case(7, 17)
```
