```python
import numpy as np
import matplotlib.pyplot as plt
import math

def exp_series(x, tol=1e-5):
    term = 1.0   # this is taking n = 0
    summation = 1.0
    
    terms = [term]
    sums = [summation]
    
    n = 1
    
    while abs(term) > tol:
        term *= x / n
        summation += term
        
        terms.append(term)
        sums.append(summation)
        
        n += 1
    
    return summation, n, np.array(terms), np.array(sums)
```

```python
exp5, n5, terms5, sums5 = exp_series(5)
expm5, nm5, termsm5, sumsm5 = exp_series(-5)

print("exp(5):", exp5, "built-in:", math.exp(5), "terms:", n5)
print("exp(-5):", expm5, "built-in:", math.exp(-5), "terms:", nm5)
```

```python
x_vals = np.arange(len(terms5))

fig, ax1 = plt.subplots()

ax1.plot(x_vals, terms5, label="Terms (x=5)") #left axis
ax1.set_xlabel("n")
ax1.set_ylabel("Term value")
ax1.grid()

ax2 = ax1.twinx() #right axis
ax2.plot(x_vals, sums5, linestyle='--', label="Sum (x=5)")
ax2.set_ylabel("Summation")

plt.title("Exponential Series (x=5)")
plt.show()
```
