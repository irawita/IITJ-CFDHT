```python
import numpy as np
import matplotlib.pyplot as plt

def q2summer(n):
    term = 2.0
    summation = 2.0
    
    terms = [term]
    sums = [summation]
    
    for k in range(1, n):
        term *= (3*k + 2) / (4*k + 1)
        summation += term
        
        terms.append(term)
        sums.append(summation)
    
    return np.array(terms), np.array(sums)
```

```python
tolerance = 1e-5
term = 2.0
summation = 2.0

k = 1

while True:
    term *= (3*k + 2) / (4*k + 1)
    
    if abs(term) < tolerance:
        break
    
    summation += term
    k += 1

print("Terms needed:", k)
```

```python
terms, sums = q2summer(60)

x = np.arange(1, len(terms)+1)

fig, ax1 = plt.subplots()


ax1.plot(x, terms, label="Individual Terms") #left axis
ax1.set_xlabel("n")
ax1.set_ylabel("Term value")
ax1.grid()

ax2 = ax1.twinx() #right axis
ax2.plot(x, sums, linestyle='--', label="Cumulative Sum")
ax2.set_ylabel("Summation")

plt.title("Series Terms and Cumulative Sum")

plt.show()
```

Computing the entire numerator and entire denominator means that with each added term, the calculated value becomes larger (as opposed to recurrent terms becoming smaller in our case). Every iteration would require dividing two very large numbers, which also on their own do not improve the fidelity of the solution.
