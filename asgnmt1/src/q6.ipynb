```python
import numpy as np
import matplotlib.pyplot as plt

def Tterm(n, x):
    return 2 * ((-1)**(n+1) / n) * np.sin(n * x)


def sumTerms(m, x):
    total = 0
    for i in range(1, m+1):
        total += Tterm(i, x)
    return total
```

```python
import math

def find_m(x, tol=1e-5):
    m = 1
    total = 0
    
    while True:
        term = Tterm(m, x)
        total += term
        
        if abs(term) < tol:
            break
        
        m += 1
    
    return total, m

x_val = np.pi / 4

approx, m = find_m(x_val)

print("Approx:", approx)
print("Exact:", x_val)
print("m required:", m)
```

```python
x_domain = np.arange(-np.pi, np.pi, 0.01)

def compute_curve(m):
    y = []
    for x in x_domain:
        y.append(sumTerms(m, x))
    return np.array(y)
```

```python
plt.figure()

# m = 1
plt.plot(x_domain, compute_curve(1), label="m=1")

# m = 2
plt.plot(x_domain, compute_curve(2), label="m=2")

# m = 3
plt.plot(x_domain, compute_curve(3), label="m=3")

# large m
plt.plot(x_domain, compute_curve(50), label="m=50 (approx)")

# actual function
plt.plot(x_domain, x_domain, linestyle='--', label="y = x")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Fourier Series Approximation of y = x")
plt.legend()
plt.grid()

plt.show()
```
